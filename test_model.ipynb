```python
import joblib
import pandas as pd

model = joblib.load("models/qoe_model_xgb_v1.0.pkl")
print("Model loaded successfully.")
```

```python
# Example: one-second snapshot from the player
sample_features = {
    "mean_vmaf": 23.9,
    "mean_psnr": 27.28,
    "mean_ssim": 0.85,
    "mean_bitrate": 148.9,
    "rebuffer_fraction": 0.0
}

import pandas as pd
input_df = pd.DataFrame([sample_features])

predicted_qoe = model.predict(input_df)[0]
print(f"Predicted QoE now: {predicted_qoe:.3f}")
```

```python
import time

for i, row in input_df.head(30).iterrows():
    features = row[["mean_vmaf","mean_psnr","mean_ssim","mean_bitrate","rebuffer_fraction"]].to_frame().T
    qoe = model.predict(features)[0]
    print(f"t={i:02d}s, predicted QoE={qoe:.3f}")
    time.sleep(1)  # simulate real-time per-second updates
```
